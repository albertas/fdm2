```python
Vaisių parduotuvė
===================
Vaisių praduotuvėje klientai gali įsigyti įvairių vaisių.

1. Sukurkite klasę `Vaisius`.
2. Sukurkite `Vaisius` klasės konstruktorių, kuris gauna vaisiaus pavadinimą, kilogramo kainą, 
kiekį parduotuvėje bei pasirinktinai kilmės šalį.
3. Sukurkite tris `Vaisius` objektus.
4. Sukurkite `Vaisius` metodą `pirkti()`, kuris gauna kiekį ir grąžina sumą, kurią reikia sumokėti. 
Jeigu parduotuvėje yra mažesnis kiekis, nei pageidaujama įsigyti - parduodamas tik turimas kiekis. 
Parduodant turimas vaisiaus kiekis turi sumažėti.
5. Sukurkite `VaisiusSuNuolaida` klasę, kuri paveldi iš `Vaisius` klasės.
6. Sukurkite `VaisiusSuNuolaida` konstruktorių, kad jis papildomai gautų ir nuolaidos dydį procentais.
7. Atnaujinkite `VaisiusSuNuolaida` metodą `pirkti`, kad į grąžinamą sumą butų įskaičiuojama ir nuolaida.
8. Sukurkite `Parduotuve` klasę ir jos konstruktorių, kuris gauna `Vaisius` objektų sąrašą.
9. Sukurkite `Parduotuve` klasės metodą `pirkti`, kuris gauna pirkinių krepšelį 
(sąrašą su vaisių pavadinimais ir kiekiais, pvz.: [('obuolys',0.5), ('bananas',0.5), ('apelsinas',1)]). 
Šis metodas grąžina visą sumą, kurią reikia sumokėti už pirkinių krepšelį bei pašalina kiekius iš parduotuvės.
10. Sukurkite `Parduotuve` objektą ir jam perduokite `Vaisius` ir `VaisiusSuNuolaida` objektų. 
Sukurtajam `Parduotuve` objektui iškvieskite `pirkti()` metodą su vaisių pavadinimais, kurie 
turi nuolaidą ir kurie neturi.

```

```python
from unittest import TestCase

class Vaisius:
    def __init__(self, pavadinimas, kilogramo_kaina, kiekis_parduotuveje, kilmes_salis=None):
        self.pavadinimas = pavadinimas
        self.kilogramo_kaina = kilogramo_kaina
        self.kiekis_parduotuveje = kiekis_parduotuveje
        self.kilmes_salis = kilmes_salis
        
    def pirkti(self, kiekis):
        if kiekis > self.kiekis_parduotuveje:
            kiekis = self.kiekis_parduotuveje
        self.kiekis_parduotuveje -= kiekis
        return kiekis * self.kilogramo_kaina
    
    
class VaisiusSuNuolaida(Vaisius):
    def __init__(self, pavadinimas, kilogramo_kaina, kiekis_parduotuveje, kilmes_salis=None, nuolaida=0):
        super().__init__(pavadinimas, kilogramo_kaina, kiekis_parduotuveje, kilmes_salis)
        self.nuolaida = nuolaida
        
    def pirkti(self, kiekis):
        moketi = super().pirkti(kiekis)
        # moketi_su_nuolaida = moketi - moketi * (self.nuolaida/100.)
        moketi_su_nuolaida = moketi * ((100 - self.nuolaida)/100.)
        return moketi_su_nuolaida
        
        
class Parduotuve:
    def __init__(self, vaisiai):
        self.vaisiai = vaisiai
        
    def pirkti(self, krepselis):
        moketi_is_viso = 0
        for pavadinimas, kiekis in krepselis:
            for vaisius in self.vaisiai:
                if vaisius.pavadinimas == pavadinimas:
                    moketi_uz_vaisiu = vaisius.pirkti(kiekis)
                    moketi_is_viso += moketi_uz_vaisiu
        return moketi_is_viso
        

class BuyFruitsMultipleTimesTestCase(TestCase):
    def test_class_vaisius_existance(self):  # Step 1.
        self.assertTrue(bool(Vaisius))
        
    def test_vaisius_instance_creation(self):  # Step 2.1
        obuolys = Vaisius('obuolys', 0.5, 100)
        self.assertEqual(obuolys.pavadinimas, 'obuolys')
        self.assertEqual(obuolys.kilogramo_kaina, 0.5)
        self.assertEqual(obuolys.kiekis_parduotuveje, 100)
        
    def test_vaisius_instance_creation_with_country(self):  # Step 2.2
        obuolys = Vaisius('obuolys', 0.5, 100, 'Lietuva')
        self.assertEqual(obuolys.pavadinimas, 'obuolys')
        self.assertEqual(obuolys.kilogramo_kaina, 0.5)
        self.assertEqual(obuolys.kiekis_parduotuveje, 100)
        self.assertEqual(obuolys.kilmes_salis, 'Lietuva')
        
    def test_serveral_vaisius_instances_creation(self):
        obuolys = Vaisius('obuolys', 0.5, 100)
        kriause = Vaisius('kriause', 1.5, 10)
        apelsinas = Vaisius('apelsinas', 1., 50)
        self.assertTrue(isinstance(obuolys, Vaisius))
        self.assertTrue(isinstance(kriause, Vaisius))
        self.assertTrue(isinstance(apelsinas, Vaisius))
        
    def test_vaisius_method_pirkti(self):
        # Vaisių parduotuvėje užtenka
        obuolys = Vaisius('obuolys', 0.5, 100)
        moketi = obuolys.pirkti(10)
        self.assertEqual(moketi, 5)
        self.assertEqual(obuolys.kiekis_parduotuveje, 90)
        
        # Vaisių parduotuvė neužtenka
        obuolys = Vaisius('obuolys', 0.5, 5)
        moketi = obuolys.pirkti(10)
        self.assertEqual(moketi, 2.5)
        self.assertEqual(obuolys.kiekis_parduotuveje, 0)
        
        # Parduodant turimas vaisiaus kiekis turi sumažėti.
        obuolys = Vaisius('obuolys', 0.5, 100)
        moketi = obuolys.pirkti(10)
        self.assertEqual(obuolys.kiekis_parduotuveje, 90)
        
    def test_class_vaisius_su_nuolaida_existance(self):  # Step 5.
        self.assertTrue(bool(VaisiusSuNuolaida))
        self.assertTrue(hasattr(VaisiusSuNuolaida, 'pirkti'))
        self.assertTrue(issubclass(VaisiusSuNuolaida, Vaisius))
        
    def test_vaisius_su_nuolaida_instance_creation(self):  # Step 6
        kivis = VaisiusSuNuolaida('kivis', 2, 20, 'Ispanija', 25)
        self.assertEqual(kivis.nuolaida, 25)
        
        kivis = VaisiusSuNuolaida('kivis', 2, 20, nuolaida=25)
        self.assertEqual(kivis.nuolaida, 25)
        
    def test_vaisius_su_nuolaida_method_pirkti(self):   # Step 7
        kivis = VaisiusSuNuolaida('kivis', 2, 20, 'Ispanija', 25)
        moketi = kivis.pirkti(10)
        self.assertEqual(moketi, 2*10*0.75)
        
    def test_parduotuve_class_existance(self):   # Step 8.1
        self.assertTrue(bool(Parduotuve))
        
    def test_parduotuve_class_existance(self):   # Step 8.2
        obuolys = Vaisius('obuolys', 0.5, 100)
        kriause = Vaisius('kriause', 1.5, 10)
        apelsinas = Vaisius('apelsinas', 1., 50)
        
        vaisiai = [obuolys, kriause, apelsinas]
        
        mindaugo_maxima = Parduotuve(vaisiai)
        
        self.assertTrue(obuolys in mindaugo_maxima.vaisiai)
        self.assertTrue(kriause in mindaugo_maxima.vaisiai)
        self.assertTrue(apelsinas in mindaugo_maxima.vaisiai)
        self.assertIn(obuolys, mindaugo_maxima.vaisiai)
        
    def test_parduotuve_method_pirkti(self):   # Step: 9
        krepselis = [('obuolys', 0.5), ('bananas',0.5), ('apelsinas',1)]
        
        obuolys = Vaisius('obuolys', 0.5, 100)
        kriause = Vaisius('kriause', 1.5, 10)
        apelsinas = Vaisius('apelsinas', 1., 50)
        
        vaisiai = [obuolys, kriause, apelsinas]
        
        mindaugo_maxima = Parduotuve(vaisiai)
        
        moketi = mindaugo_maxima.pirkti(krepselis)
        self.assertEqual(moketi, 0.5 * 0.5 + 0 + 1 * 1)
        
        self.assertEqual(obuolys.kiekis_parduotuveje, 100 - 0.5)
        
    def test_parduotuve_method_pirkti_with_various_vaisius_instances(self):   # Step: 10
        krepselis = [('obuolys', 0.5), ('bananas',0.5), ('apelsinas',1), ('kivis', 1), ('persimonas', 2)]
        
        obuolys = Vaisius('obuolys', 0.5, 100)
        kriause = Vaisius('kriause', 1.5, 10)
        apelsinas = Vaisius('apelsinas', 1., 50)
        kivis = VaisiusSuNuolaida('kivis', 1., 50, nuolaida=25)
        persimonas = VaisiusSuNuolaida('persimonas', 1., 1, nuolaida=50)
        
        vaisiai = [obuolys, kriause, apelsinas, kivis, persimonas]
        
        mindaugo_maxima = Parduotuve(vaisiai)
        
        moketi = mindaugo_maxima.pirkti(krepselis)
        self.assertEqual(moketi, 0.5 * 0.5 + 0 + 1 * 1 + 0.75 + 0.5)
        
        self.assertEqual(obuolys.kiekis_parduotuveje, 100 - 0.5)
        
  
    
    
import unittest
unittest.main(argv=[''], exit=False)
```
